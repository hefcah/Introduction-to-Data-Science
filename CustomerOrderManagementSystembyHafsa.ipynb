```python
# Dictionary to store customer orders
customer_orders = {}
total_sales = 0             #since order has not been placed so I've iitialized it eith zero,will be updated later on
total_customers = 0         #ince no customer has come till now , will be updated later

bill = 0
cont = 'C' or 'c'     #within few lines,we willl be using continue statement

while cont == 'C' or cont == 'c':
    a = input("Enter Your Name ")

    # Check if the customer has purchased previously
    if a in customer_orders:
        print("Welcome back,", a)
    
    b = input("Do you want frozen yogurt or ice -cream ?")
    if b == 'Y' or b == 'y':
        print("Yogurt")
        bill = bill + 200      #before asking size i am adding bill because price of ice-cream and yogurt is not same
    elif b == 'I' or b == 'i':
        print("Ice-Cream")
        bill = bill + 120       #if bill was not added it would give same prices for yogurt and ice-cream 
    else:
        print("Choice Invalid")
    
    c = input("Enter size for desired item ")
    if c == 'S' or c == 's':
        print("Small")
        bill = bill + 30
    elif c == 'M' or c == 'm':
        print("Medium")
        bill = bill + 60
    elif c == 'L' or c == 'l':
        print("Large")
        bill = bill + 100
    
    d = input("Enter the type of container you want ")
    if d == 'V' or d == 'v':
        print("Cone")
        bill = bill + 50
    elif d == 'U' or d == 'u':
        print("Cup")
        bill = bill + 30
    else:
        print("Non-Valid Entry")
    
    e = input("Would you like to add topping ?")
    if e == 'Y' or e == 'y':
        print("Yes")
        f = input("What topping do you like ? ")
        if f == 'G' or f == 'g':
            print("Gummies")
            bill = bill + 50
        elif f == 'N' or f == 'n':
            print("Nuts")
            bill = bill + 80
    elif e == 'N' or e == 'n':
        print("No")
    
    j = input("Do you like to have cookies ? ")
    if j == 'Y' or j == 'y':
        print("Yes")
        q = input("What kind of Cookies ? ")
        if q == 'R' or q == 'r':
            print("Raisin")
            bill = bill + 100
        elif q == 'C' or q == 'c':
            print("Chocolate")
            bill = bill + 120
        h = input("Enter size of cookies ")
        if h == 'R' or h == 'r':
            print("Regular")
            bill = bill + 50
        elif h == 'L' or h == 'l':
            print("Large")
            bill = bill + 70
        elif j == 'N' or j == 'n':
            print("No")
        else:
            print("Non-Valid input")
    
    w = int(input("What day of week is this ? "))
    if w == 6 or w == 7:          #6 for saturday,7 for sunday
        print("Weekend Special discount")
        bill = bill - bill * 10 // 100       #formulae to calculate 10% discount

    # Store customer name and order total in the dictionary
    customer_orders[a] = bill
    total_sales += bill
    total_customers += 1

    print("Your bill for the purchase is ", bill)
    cont = input("Do you want to continue ?")
    if cont != 'C' and cont != 'c':
        break

# Print the customer orders dictionary
print("\nCustomer Orders:")
for customer, total in customer_orders.items():
    print(f"{customer}: {total}")

# Calculate and print average customer orders
average_orders = total_sales / total_customers if total_customers > 0 else 0   
#for the above line i was not getting correct indentation so i refered gpt for this
print("\nAverage Customer Orders:", average_orders)
```
